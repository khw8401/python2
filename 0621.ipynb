```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

```python
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
cancer.data, cancer.target, stratify= cancer.target, random_state=42)
tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train, y_train)
print("훈련 세트 정확도: {:.3f}".format(tree.score(X_train, y_train)))
print("테스트 세트 정확도: {:.3f}".format(tree.score(X_test, y_test)))
```

```python
tree = DecisionTreeClassifier(max_depth=4, random_state=0)
tree.fit(X_train, y_train)
print("훈련 세트 정확도: {:.3f}".format(tree.score(X_train, y_train)))
print("테스트 세트 정확도: {:.3f}".format(tree.score(X_test, y_test)))
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file="tree.dot", class_names=["악성","양성"],
               feature_names=cancer.feature_names, impurity=False, filled=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
ran_price = pd.read_csv("ram_price.csv")
plt.semilogy(ran_price.date, ran_price.price)
```

```python
plt.plot(ran_price.date, ran_price.price)
```

```python
from sklearn.tree import DecisionTreeRegressor
# 2000년 이전을 훈련 데이터, 2000년 이후를 테스트 데이터
data_train = ran_price[ran_price.date < 2000]
data_test = ran_price[ran_price.date >= 2000]
```

```python
data_train.shape
```

```python
data_test.shape
```

```python
X_train =  data_train.date.to_numpy()
X_train
```

```python
import numpy as np
X_train = X_train[:,np.newaxis]
X_train
```

```python
y_train = np.log(data_train.price)
y_train
```

```python
dt = DecisionTreeRegressor().fit(X_train, y_train)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression().fit(X_train, y_train)
```

```python
X_all = ran_price.date.to_numpy()[:,np.newaxis]
```

```python
pred_dt = dt.predict(X_all)
pred_lr = lr.predict(X_all)
```

```python
price_dt = np.exp(pred_dt)
price_lr = np.exp(pred_lr)
```

```python
plt.semilogy(data_train.date, data_train.price, label="train data")
plt.semilogy(data_test.date, data_test.price, label="test data")
plt.semilogy(ran_price.date, price_dt, label="pred of dt")
plt.semilogy(ran_price.date, price_lr, label="pred of lr")
plt.legend()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=100, noise=0.25, random_state=0)
X_train, X_test, y_train, y_test = train_test_split(
X,y, stratify= y, random_state=42)
forest = RandomForestClassifier(n_estimators=5, random_state=2).fit(X_train, y_train)
fig, axes = plt.subplots(2,3,figsize = (20,10))
for i, (ax,tree) in enumerate(zip(axes.ravel(),forest.estimators_)):
    mglearn.plots.plot_tree_partition(X,y,tree,ax=ax)

mglearn.plots.plot_2d_separator(forest, X, fill=True, alpha = 0.4)
mglearn.discrete_scatter(X[:,0],X[:,1],y)
```
