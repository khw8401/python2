```python
import mglearn
```

```python
X,y = mglearn.datasets.load_extended_boston()
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X_train, X_test,y_train,y_test = train_test_split(X,y,random_state=0)
lr = LinearRegression().fit(X_train, y_train)
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge().fit(X_train, y_train)
ridge10 = Ridge(alpha=10).fit(X_train, y_train)
ridge01 = Ridge(alpha=0.1).fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
plt.plot(ridge.coef_,"^", label='alpha1')
plt.plot(ridge10.coef_,"s", label='alpha10')
plt.plot(ridge01.coef_,"^", label='alpha0.1')
# plt.plot(lr.coef_, 'o',label = "Linear")
plt.legend()
```

```python
from sklearn.linear_model import Lasso
Lasso = Lasso().fit(X_train,y_train)
```

```python
Lasso.score(X_train, y_train)
```

```python
Lasso.score(X_test, y_test)
```

```python
import numpy as np
np.sum(Lasso.coef_!= 0)
```

```python
Lasso001 = Lasso(alpha=0.01, max_iter = 50000).fit(X_train, y_train)
```

```python
Lasso00001 = Lasso(alpha=0.0001, max_iter = 50000).fit(X_train, y_train)
```

```python
plt.plot(Lasso.coef_, 's')
plt.plot(Lasso001.coef_,'^')
plt.plot(Lasso00001.coef_,'v')
```
