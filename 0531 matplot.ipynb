```python
m = np.random.standard_normal((500,2))
r = np.sum(m**2, axis=1)
plt.figure(figsize=(5,5))
plt.ylim(-5,5)
plt.xlim(-5,5)
plt.scatter(m[:,0],m[:,1], c='0.7', marker='s',
           s=32*r, edgecolor='k')
plt.show()
```

```python
import matplotlib.patches as patches
ax = plt.axes(polar=True)

radius = [1,0.5,0.7,0.3,0.1,0.2,0.6,0.8,0.5]
theta = np.linspace(0, 2*np.pi,len(radius),endpoint=False)
points = np.vstack((theta,radius)).T
plt.gca().add_patch(patches.Polygon(points,color='0.5'))
plt.show()
```
