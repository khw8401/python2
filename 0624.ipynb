```python
### Clustering
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import mglearn
```

```python
## K-Means
X, y = make_blobs(random_state = 1, cluster_std =3)
mglearn.discrete_scatter(X[:,0], X[:,1]);
```

```python
kmeans = KMeans(n_clusters=3).fit(X)
mglearn.discrete_scatter(X[:,0], X[:,1], kmeans.labels_); # predict 메서드를 사용하는 것과 동일 
```

```python
mglearn.discrete_scatter(X[:,0], X[:,1], y); 
```

```python
# 알고리즘이 실패하는 경우(정상)
X, y = make_blobs(n_samples = 200, random_state = 170, cluster_std =[1, 2.5, 0.5])
mglearn.discrete_scatter(X[:,0], X[:,1], y);
```

```python
# 알고리즘이 실패하는 경우(실패)
y_pred = KMeans(n_clusters=3).fit_predict(X)      # kmeans 특성상 균등하게 기준은 갯수가 아니라 거리에 따른 구분이다.
mglearn.discrete_scatter(X[:,0], X[:,1], y_pred); # 비지도학습이기 때문에 지도 학습에서 진행하는 예측에 오류가 생기는 것
```

```python
# 알고리즘이 실패하는 경우(실패2)
import numpy as np
X, y = make_blobs(random_state=170, n_samples = 600)
rng = np.random.RandomState(74)
transformation = rng.normal(size=(2,2))
X = np.dot(X, transformation)                # dot의 유무로 그래프가 바뀜 
mglearn.discrete_scatter(X[: , 0], X[: , 1]);
```

```python
kmeans = KMeans(n_clusters=3).fit(X)
y_pred = kmeans.predict(X)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y_pred);
mglearn.discrete_scatter(kmeans.cluster_centers_[:, 0],
                         kmeans.cluster_centers_[:, 1],
                        markers='s') # 중점
```

```python
kmeans.cluster_centers_
# 중점을 중심으로 비지도학습의 일부로 원형이 아닌 클러스터를 구분하지 못하는 k-means
```

```python
# 알고리즘이 실패하는 경우(실패3)
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=200, noise = 0.05, random_state = 0)

model = KMeans(n_clusters=2).fit(X)
y_ = model.predict(X)

mglearn.discrete_scatter(X[: , 0], X[: , 1], y)
```

```python
mglearn.discrete_scatter(X[: , 0], X[: , 1], y_)
```

```python
mglearn.discrete_scatter(X[: , 0], X[: , 1], y_)
mglearn.discrete_scatter(model.cluster_centers_[:, 0],
                         model.cluster_centers_[:, 1],
                         markers='s', s = '30'); # 중점
```

```python
from sklearn.cluster import AgglomerativeClustering
X, y = make_blobs(random_state=1)
agg_model = AgglomerativeClustering()
y_ = agg_model.fit_predict(X)
mglearn.discrete_scatter(X[: , 0], X[: , 1], y_)
```

```python
from sklearn.cluster import DBSCAN
```

```python
X,y = make_blobs(random_state = 0, n_samples=12)
mglearn.discrete_scatter(X[:,0],X[:,1],y)
```

```python
dbscan = DBSCAN(eps=2)
y_ = dbscan.fit_predict(X)
mglearn.discrete_scatter(X[:,0],X[:,1],y_)
```

```python
X, y = make_moons(n_samples=200, noise=0.05, random_state=0)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X_ = scaler.transform(X)
dbscan = DBSCAN()
y_ = dbscan.fit_predict(X_)
mglearn.discrete_scatter(X_[:,0],X_[:,1],y)
import matplotlib.pyplot as plt
plt.show()
```
