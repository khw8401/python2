```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.hist(np.random.randn(100), bins=20, color = 'm', alpha = 0.3) # alpha 투명도

ax2 = fig.add_subplot(2,2,2)
ax2.scatter(np.arange(30),np.arange(30)+3*np.random.randn(30),color = 'g') # scatter는 x,y축 필요 앞 30 x , 뒤 30+3*np 는 y 

ax3 = fig.add_subplot(2,2,3)
plt.plot(np.random.randn(50).cumsum(),'y--')
```
