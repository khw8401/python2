```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
import mglearn
```

```python
X,y = make_moons(n_samples = 100, noise = 0.25, random_state = 3)
X_train, X_test, y_train, y_test = train_test_split(X,y,stratify = y, random_state=42)
mlp = MLPClassifier(solver='lbfgs', random_state=0).fit(X_train, y_train)
mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
```

```python
X,y = make_moons(n_samples = 100, noise = 0.25, random_state = 3)
X_train, X_test, y_train, y_test = train_test_split(X,y,stratify = y, random_state=42)
mlp = MLPClassifier(solver='lbfgs', random_state=0, hidden_layer_sizes=[10],max_iter=1000) # max_iter - 실수 할 기회
mlp.fit(X_train, y_train)
mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
```

```python
X,y = make_moons(n_samples = 100, noise = 0.25, random_state = 3)
X_train, X_test, y_train, y_test = train_test_split(X,y,stratify = y, random_state=42)
mlp = MLPClassifier(solver='lbfgs', random_state=0, hidden_layer_sizes=[10,10]) # max_iter - 실수 할 기회
mlp.fit(X_train, y_train)
mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
```

```python
X,y = make_moons(n_samples = 100, noise = 0.25, random_state = 3)
X_train, X_test, y_train, y_test = train_test_split(X,y,stratify = y, random_state=42) # tanh 더 매끄러운 결정체를 원할 떄 사용
mlp = MLPClassifier(solver='lbfgs', random_state=0, activation= 'tanh',hidden_layer_sizes=[10,10])
mlp.fit(X_train, y_train)
mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2,5, figsize=(25,8))
for ax_y, n_nodes in zip (axes, [10,100]):
    for ax_x, alpha in zip(ax_y, [0.0001, 0.01, 0.1, 1, 10]):
        mlp = MLPClassifier(solver='lbfgs', random_state=0, alpha= alpha, hidden_layer_sizes=[n_nodes, n_nodes])
        mlp.fit(X_train, y_train)
        mglearn.plots.plot_2d_separator(mlp, X_train, ax=ax_x)
        mglearn.discrete_scatter(X_train[:,0], X_train[:,1],y_train,ax=ax_x)
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
cancer.data.max(axis=0)
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer.data,cancer.target, random_state=0)
```

```python
mlp = MLPClassifier(random_state=42).fit(X_train, y_train)
```

```python
mlp.score(X_train, y_train)
```

```python
mlp.score(X_test, y_test)
```

```python
train_mean = X_train.mean(axis=0)
train_mean
```

```python
train_std = X_train.std(axis=0)
train_std
```

```python
X_train_s = X_train - train_mean/train_std
X_train_s
```

```python
X_test_s = X_test - train_mean/train_std
X_test_s
```

```python
mlp = MLPClassifier(random_state=42).fit(X_train_s, y_train)
```

```python
mlp.score(X_train_s, y_train)
```

```python
mlp.score(X_test_s, y_test)
```

```python
mlp = MLPClassifier(random_state=0, max_iter=1000).fit(X_train_s, y_train)
```

```python
mlp.score(X_train_s, y_train)
```
